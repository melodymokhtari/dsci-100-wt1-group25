In the dataset, “players.csv”, there are 196 observations and 9 columns, two in which are hard to read. 
- “experience” (object), player’s experience with the game
- “subcribe” (boolean), did they subscribe?
- “hashedEmail” (object), the email of the player
- “played_hours” (float), hours played in game
- “name” (object), name of player
- “gender” (object), gender of player
- “age” (integer), age of player
- individualId and organizationName are hard to read values (these variables will be ignored)

Can we predict a player's likelihood of subscribing to the newsletter based on their personal and in-game characteristics?

- We can use predictive variables (age, gender (turned binary), played hours, experience (turned binary) from “players.csv” for defining the characteristics.
- We can wrangle the data to only have the variables we are interested in. This is when we'll drop columns that won't be useful, like "individualId" and "organizationName".
- We can change the variables that have object observations to numbers so we can use classification.
- Class to prediction: subcribed or not?

Below are minimal wranglings for the dataset.

```python
url = "https://drive.google.com/file/d/1Mw9vW0hjTJwRWx0bDXiSpYsO3gKogaPz/view?usp=sharing"
players=pd.read_csv("https://drive.google.com/uc?export=download&id="+ url.split('/')[-2])
dropped_players=players.drop(columns=["individualId","organizationName","hashedEmail"])
dropped_players
```

```python
players_plot_scatter = alt.Chart(dropped_players).mark_point().encode(
    x = alt.X("age:Q").title("Age of Player"),
    y= alt.Y("played_hours").title("Played Hours"),
    color = alt.Color("experience").title("Experience").scale(scheme = "set1")
)
players_plot_scatter
```

From players_plot_scatter, we can see how the age of the player compares to their played hours. And, the different categories of experience are coloured to see how experience can be compared. 

```python
players_plot_bar2 = alt.Chart(dropped_players).mark_bar().encode(
    x = alt.X("subscribe").title("Players Subcribed"),
    y = alt.Y("played_hours").title("Hours Played"),
    color = alt.Color("experience").title("Experience").scale(scheme = "set1")
)
players_plot_bar2
```

From players_plot_bar2, we can see how when the players are subcribed, how are their playing hours reflected and what is their experience.

```python
players_plot_bar3 = alt.Chart(dropped_players).mark_bar().encode(
    x = alt.X("experience").title("Experience of Players"),
    y = alt.Y("played_hours").title("Played Hours"),
    color = alt.Color("subscribe").title("Subscribed").scale(scheme = "set1")
)
players_plot_bar3
```

From players_plot_bar3, we can see how the experience of players compares to their playing hours and see how these two variables come together to see whether the players are subscribed or not.

To address the question “Can we predict a player's likelihood of subscribing to the newsletter based on their personal and in-game characteristics?”I propose using a classification approach using the dataset “players.csv”.

The variables “experience” and "subscribe" are originally categorical, but they can be converted to numerical form to allow for efficient modeling and analysis. Although encoded numerically, the outcome remains categorical (subscriber vs. non-subscriber), making classification possible for this dataset. This allows us to use KNN and have predictors help us find the categorial result. 

Classification is ideal for predicting numerical outcomes and provides interpretable insights into how player characteristics such as experience, age, gender, and played_hours affect the likelihood of subscription.

Assumptions:
- The outcome variable is binary (0 or 1).
- Observations are independent.
- Predictors values are even.

Potential limitations:
- Sensitive to outliers and how categorical variables are encoded.
- Can underperform compared to more flexible models if complex patterns exist in the data.

Model performance will be performed by using metrics such as accuracy, precision, recall and K-neighbours Classifier. Data will be scaled, clean, and continuous variables standardized as needed. We will wrangle the data by dropping variables that we are not using and unreadable. These variables will be "individualId","organizationName" and "hashedEmail", as these variables are not needed as they are not "characteristics". The dataset will be split into training (70%) and testing (30%) sets, with a 5 cross-validation on the training data to validate the model.
